```python
df_d['rating']=df_d['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
df_d['reviewText']=df_d['reviewText'].str.lower()
```

```python
df_d.head()
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
import re
from bs4 import BeautifulSoup

# 1. Remove special characters/keep only alphanumeric
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: re.sub('[^a-z A-Z 0-9]+', ' ', x))

# 2. Remove stopwords (assuming 'stopwords' list is predefined)
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: " ".join([y for y in x.split() if y not in stopwords.words('english')]))

# 3. Remove URLs (http, https, ftp, ssh)
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', x))

# 4. Remove HTML tags using BeautifulSoup
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())

# 5. Remove extra oral spaces
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df_d.head()
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
nltk.download('wordnet')
```

```python
def lemmatize(text):
  return " ".join([lem.lemmatize(word) for word in text.split()])
```

```python
df_d['reviewText'] = df_d['reviewText'].apply(lambda x: lemmatize(x) )
```

```python
df_d.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_d['reviewText'],df_d['rating'],test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x_train_b=cv.fit_transform(x_train).toarray()
x_test_b=cv.transform(x_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x_train_t=tfidf.fit_transform(x_train).toarray()
x_test_t=tfidf.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model=GaussianNB()
bow=nb_model.fit(x_train_b,y_train)
ttfidf=nb_model.fit(x_train_t,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
y_pred_cv=bow.predict(x_test_b)
y_pred_train_cv=bow.predict(x_train_b)
```

```python
y_pred_tf=bow.predict(x_test_t)
y_pred_train_tf=bow.predict(x_train_t)
```

```python
print(accuracy_score(y_train,y_pred_train_cv))
print(accuracy_score(y_test,y_pred_cv))
```

```python
print(accuracy_score(y_train,y_pred_train_tf))
print(accuracy_score(y_test,y_pred_tf))
```
