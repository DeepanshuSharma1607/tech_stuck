## Reshaping stacking ,  squeezing tensors

* Reshaping - Reshape an input tesor
* Stacking - Combine multiple tensors on top of each other vstack and hstack
* Squeez - Remove all `1` dimension from tensor
* Permuate - Return a view of the input with dimensions permuted(swapped)
* View - Return a view of the input tensor of certain shape but keeps the same memory as orignal tensor

```python
import torch
```

```python
x=torch.arange(1.,10.)
x,x.shape
```

```python
x_reshaped=x.reshape(1,9)
x_reshaped,x_reshaped.shape
```

```python
x_reshaped=x.reshape(9,1)
x_reshaped,x_reshaped.shape
```

```python
x=torch.arange(1.,13.)
x,x.shape
```

```python
x_reshaped=x.reshape(2,6)
x_reshaped,x_reshaped.shape
```

```python
x_reshaped=x.reshape(3,4)
x_reshaped,x_reshaped.shape
```

```python
z=x.view(1,12)
z,z.shape
# change in x will change z and vise versa
```

```python
z[:,0]=5
z,x
```

```python
x_stacked=torch.stack([x,x,x,x],dim=1)
x_stacked
```

```python
x_stacked=torch.vstack([x,x,x,x])
x_stacked,x_stacked.shape
```

```python
x_stacked=torch.hstack([x,x,x,x])
x_stacked,x_stacked.shape
```
