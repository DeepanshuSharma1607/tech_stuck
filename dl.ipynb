```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "diabetes.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "mathchi/diabetes-data-set",
  file_path,
  # Provide any additional arguments like
  # sql_query or pandas_kwargs. See the
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
X=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression,Perceptron
p=LogisticRegression(max_iter=100000)
p.fit(X_train,y_train)
print(p.coef_)
print(p.intercept_)
'''
perceptron : [[ 530.     168.    -311.      13.     102.     -81.2      5.978   70.   ]]
[-149.]

logistic regression : [[ 0.06106262  0.0337888  -0.01387792  0.00474554 -0.0019446   0.10151985
   0.59002294  0.03768787]]
[-8.93569889]

'''
```

```python
from sklearn.metrics import accuracy_score
y_pred_test=p.predict(X_test)
y_pred_train=p.predict(X_train)
print('accuracy_of_training ',accuracy_score(y_pred_train,y_train))
print('accuracy_of_testing ',accuracy_score(y_pred_test,y_test))

'''
perceptron : accuracy_of_training  0.5211726384364821
accuracy_of_testing  0.5714285714285714


logistic regression :
accuracy_of_training  0.7703583061889251
accuracy_of_testing  0.7402597402597403

'''
```

```python
def perceptron(x,y):
  x=np.insert(x,0,1,axis=1)
  w=np.ones(x.shape[1])
  alpha=0.01
  for i in range(10000):
    j=np.random.randint(0,x.shape[0])
    y_hat=get(np.dot(x[j],w))
    # Corrected: Use .iloc for positional indexing on the Pandas Series y
    w=w+alpha*(y.iloc[j]-y_hat)*x[j]

  return w[0],w[1:]
```

```python
def get(x):
  return 1 if x>0 else 0
```

```python
intercept_,coef_=perceptron(X_train,y_train)
print(coef_,intercept_)
```

```python
y_pred_train_=np.dot(X_train,coef_)+intercept_
y_pred_test_=np.dot(X_test,coef_)+intercept_

vec_get=np.vectorize(get)
y_pred_train_=vec_get(y_pred_train_)
y_pred_test_=vec_get(y_pred_test_)

print('accuracy_of_training ',accuracy_score(y_pred_train_,y_train))
print('accuracy_of_testing ',accuracy_score(y_pred_test_,y_test))
```

```python
def perceptron_(x,y):
  w=np.ones(x.shape[1])
  b=1
  alp=0.1
  for i in range(1000):
    j=np.random.randint(0,x.shape[0])
    x_=x.iloc[j].values
    y_=y.iloc[j]
    z=np.dot(x_,w)+b
    if z*y_<0:
      w=w+alp*y_*x_
      b=b+alp*y_*x_

  return w,b
```
